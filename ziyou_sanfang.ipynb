```python
import pandas as pd
from pyhive import hive
from matplotlib import pyplot as plt
host='*****'
username='*****'
password='*****'
port=10000
data_base_name = '*****'
conn1 = hive.Connection(host=host,
                       port=port,
                       auth="CUSTOM",
                       database=data_base_name,
                       username=username,
                       password=password)
sql_station_id="select station_id from ***** where asset_type = 1 and dt = '2022-06-19'"
station_id = pd.read_sql(sql_station_id, conn1)
# conn.close
```

```python
print(station_id)
```

```python
predict_data = pd.read_csv("C:/Users/IdeaProjects/final/feature_data_rf_predict_train_data.csv",encoding = "utf_8_sig")
```

```python
predict_data.head()
```

```python
res = []
for i, val in station_id["station_id"].items():
    res.append(val)
index = []
ans = []
for i, val in predict_data['station_id'].items():
    if val in res:
        index.append(i)
    else:
        ans.append(i)
ziyou_predict = predict_data.iloc[index]
sanfang_predict = predict_data.iloc[ans]
ziyou_predict.head()
sanfang_predict.head()
```

```python
import bisect
ziyou_predict.head()
ziyou_val = list(ziyou_predict["rf_pred"])
sorted_ziyou_val = sorted(ziyou_val)
ziyou_indexs = [(len(ziyou_val)+1 - bisect.bisect(sorted_ziyou_val, i)) for i in ziyou_val]
ziyou_predict_indexs = pd.DataFrame(ziyou_indexs, columns=["indexs"])


sanfang_val = list(sanfang_predict["rf_pred"])
sorted_sanfang_val = sorted(sanfang_val)
sanfang_indexs = [(len(sanfang_val) +1 -bisect.bisect(sorted_sanfang_val, i)) for i in sanfang_val]
sanfang_predict_indexs = pd.DataFrame(sanfang_indexs, columns=["indexs"])
ziyou_predict_indexs.to_csv("C:/Users/IdeaProjects/final/ziyou_predict_indexs.csv",encoding = "utf_8_sig")
sanfang_predict_indexs.to_csv("C:/Users/IdeaProjects/final/sanfang_predict_indexs.csv",encoding = "utf_8_sig")

```
